```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Simulate a time-varying stimulus
np.random.seed(42)
n_timepoints = 1000
stimulus = np.random.randn(n_timepoints)  # Random noise stimulus

# Simulate a neuron's temporal receptive field (TRF)
true_trf = np.exp(-np.linspace(0, 1, 50)) * np.cos(np.linspace(0, 4 * np.pi, 50))

# Create a simulated response by convolving stimulus with the true TRF
response = np.convolve(stimulus, true_trf, mode='full')[:n_timepoints] + 0.1 * np.random.randn(n_timepoints)

# Plot the true TRF and the simulated response
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(true_trf)
plt.title('True TRF (Neural Filter)')
plt.xlabel('Time lag')
plt.ylabel('Response strength')

plt.subplot(1, 2, 2)
plt.plot(response)
plt.title('Simulated Neuronal Response')
plt.xlabel('Time')
plt.ylabel('Response')
plt.tight_layout()
plt.show()

# Estimate TRF via reverse correlation
# We'll use 50 time lags to estimate the TRF
n_lags = 50
receptive_field = np.zeros(n_lags)

# Reverse correlation: slide the stimulus across the response at different lags
for lag in range(n_lags):
    receptive_field[lag] = np.dot(stimulus[:n_timepoints - lag], response[lag:])

# Normalize by the number of points (optional for consistency)
receptive_field /= np.linalg.norm(receptive_field)

# Plot the estimated TRF compared to the true TRF
plt.figure(figsize=(8, 4))
plt.plot(true_trf, label="True TRF", linestyle="--", color="blue")
plt.plot(receptive_field, label="Estimated TRF", color="red")
plt.title('Estimated TRF vs. True TRF')
plt.xlabel('Time lag')
plt.ylabel('Receptive field strength')
plt.legend()
plt.show()
```
